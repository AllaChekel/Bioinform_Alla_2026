##3.2



**ход работы**

1)Сначала транскриптные последовательности были переведены в белковые с помощью утилиты TransDecoder.LongOrfs.Результатом его работы является файл longest_orfs.pep, который затем можно использовать для дальнейших гомологичных поисков, в том числе с помощью BLASTP.

2)Для работы были использованы следующие команды:

TransDecoder.LongOrfs -t Photinus_pyralis_gca008802855v1rs.Ppyr1.3.cdna.all.fa
TransDecoder.LongOrfs -t ILUMI_OGS1.2.mRNA.fa

3) После этого были получены файлы:

Photinus_pyralis_gca008802855v1rs.Ppyr1.3.cdna.all.fa.transdecoder_dir/longest_orfs.pep
ILUMI_OGS1.2.mRNA.fa.transdecoder_dir/longest_orfs.pep
4)Затем к идентификаторам белков были добавлены префиксы, чтобы различать, из какого вида происходит каждый кандидат, после чего белковые последовательности были объединены в один общий FASTA-файл:

sed 's/^>/>Ppyr|/' Photinus_pyralis_gca008802855v1rs.Ppyr1.3.cdna.all.fa.transdecoder_dir/longest_orfs.pep > Ppyr.longest_orfs.pep
sed 's/^>/>Ilumi|/' ILUMI_OGS1.2.mRNA.fa.transdecoder_dir/longest_orfs.pep > Ilumi.longest_orfs.pep
cat Ppyr.longest_orfs.pep Ilumi.longest_orfs.pep > beetle_transdecoder_all.pep

5)На основе этого объединенного файла была построена локальная белковая BLAST-база с помощью makeblastdb. Утилита makeblastdb создает BLAST-базы из FASTA-файлов; в работе была использована белковая база (-dbtype prot).makeblastdb -in beetle_transdecoder_all.pep -dbtype prot -parse_seqids -blastdb_version 4 -out beetle_transdecoder_db

6)Далее был выполнен поиск гомологов люциферазы с помощью blastp, так как и запрос, и база были белковыми:

blastp -query Luciola_cruciata_luciferase.faa -db beetle_transdecoder_db -evalue 1e-5 -outfmt "6 qseqid sseqid pident l

7) Из полученной таблицы были выбраны 10 лучших уникальных хитов, после чего их идентификаторы были сохранены в отдельный файл:

sort -k7,7gr -k6,6g luciferase_vs_beetles.tsv | awk '!seen[$2]++' | head -n 10 > top10_luciferase_hits.tsv
cut -f2 top10_luciferase_hits.tsv > top10_ids.txt

8)На последнем этапе требуемые 10 последовательностей были сохранены в новый FASTA-файл с помощью утилиты seqtk.

seqtk subseq beetle_transdecoder_all.pep top10_ids.txt > top10_luciferase_candidates.fasta

В результате были получены файлы:

top10_luciferase_hits.tsv — таблица 10 лучших хитов
top10_ids.txt — список их идентификаторов
top10_luciferase_candidates.fasta — итоговый FASTA-файл с 10 лучшими кандидатами на последовательность люциферазы

Белковая последовательность люциферазы Luciola cruciata была использована как запрос в blastp, что позволило найти наиболее похожие белковые последовательности в анализируемых транскриптомах.
